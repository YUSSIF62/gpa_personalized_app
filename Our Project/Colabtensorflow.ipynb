```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split

# Step 1: Load the datasets
student_data_path = r"C:\Users\Abdul-Rahaman Yussif\Desktop\Our Project\student_study_habits_dataset.csv"
courses_data_path = r"C:\Users\Abdul-Rahaman Yussif\Desktop\Our Project\courses_by_semester_year.csv"

# Load the student data
student_df = pd.read_csv(student_data_path)
# Load the course offerings data
courses_df = pd.read_csv(courses_data_path)

# Step 2: Preprocess the student data
# Encode categorical features: study_habits and semester
encoder = OneHotEncoder()
encoded_study_habits = encoder.fit_transform(student_df[['study_habits', 'semester']]).toarray()

# Normalize numerical features: current_cgpa and target_cgpa
scaler = StandardScaler()
scaled_cgpa = scaler.fit_transform(student_df[['current_cgpa', 'target_cgpa']])

# Combine all features into a single array for the student features
student_features = np.hstack((scaled_cgpa, encoded_study_habits, student_df[['level']].values))

# Step 3: Create labels based on timetable slots using course offerings
# For simplicity, we will assign a numeric label to each combination of level and semester
def encode_timetable_slots(row):
    level, semester = row['level'], row['semester']
    # Get the courses for the given level and semester from the courses dataset
    courses = courses_df[(courses_df['level'] == level) & (courses_df['semester'] == semester)]['courses'].values[0]
    return courses

# Create a new column in the student dataset with encoded timetable slots
student_df['timetable_slots'] = student_df.apply(encode_timetable_slots, axis=1)

# Convert the timetable slot descriptions to numerical labels for training
timetable_slots = student_df['timetable_slots'].astype('category')
labels = timetable_slots.cat.codes
num_classes = len(timetable_slots.cat.categories)

# Step 4: Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(student_features, labels, test_size=0.2, random_state=42)

# Step 5: Define the neural network model using the student features
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(student_features.shape[1],)),  # Ensure student_features is defined before this step
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dropout(0.3),  # Dropout layer to prevent overfitting
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(num_classes, activation='softmax')  # Softmax for classification
])

# Step 6: Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Step 7: Train the model
model.fit(X_train, y_train, epochs=50, batch_size=8, validation_data=(X_test, y_test))

# Step 8: Save the trained model for future use
model.save('student_timetable_model.h5')
print("Model training completed and saved as 'student_timetable_model.h5'.")
```
